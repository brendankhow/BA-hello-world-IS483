```python
# Heuristic: columns with only {0,1} (ignoring NaNs)
binary_cols = []
for c in df.columns:
    vals = pd.unique(df[c].dropna())
    if len(vals) > 0 and len(vals) <= 3:
        s = set(map(lambda x: x if isinstance(x, (int, float)) else str(x).strip(), vals))
        if s.issubset({0, 1, 0.0, 1.0, True, False}):
            binary_cols.append(c)

issue_counts = (df[binary_cols]
                .fillna(0)
                .astype(float)
                .sum()
                .sort_values(ascending=False)
                .head(10))

plt.figure(figsize=(10,6))
sns.barplot(x=issue_counts.values, y=issue_counts.index)
plt.title("Top 10 Most Frequently Checked Issue Categories")
plt.xlabel("Count (sum of 1s)")
plt.ylabel("Issue Category")
plt.show()

issue_counts
```

```python
def phq9_severity(x):
    if pd.isna(x): return None
    if 0 <= x <= 4: return "Minimal"
    if 5 <= x <= 9: return "Mild"
    if 10 <= x <= 14: return "Moderate"
    if 15 <= x <= 19: return "Moderately Severe"
    if x >= 20: return "Severe"

def gad7_severity(x):
    if pd.isna(x): return None
    if 0 <= x <= 4: return "Minimal"
    if 5 <= x <= 9: return "Mild"
    if 10 <= x <= 14: return "Moderate"
    if x >= 15: return "Severe"

df["PHQ9_Severity"] = df["PHQ-9 (Int)"].apply(phq9_severity)
df["GAD7_Severity"] = df["GAD-7 (Int)"].apply(gad7_severity)

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="PHQ9_Severity", order=["Minimal","Mild","Moderate","Moderately Severe","Severe"])
plt.title("PHQ-9 Severity Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="GAD7_Severity", order=["Minimal","Mild","Moderate","Severe"])
plt.title("GAD-7 Severity Distribution")
plt.show()
```

```python
phq_by_year = pd.crosstab(df["Year"], df["PHQ9_Severity"], normalize="index")*100
phq_by_year.plot(kind="bar", stacked=True, figsize=(10,6))
plt.title("PHQ-9 Severity Distribution by Year of Study (%)")
plt.ylabel("Percentage")
plt.show()

gad_by_gender = pd.crosstab(df["M/F"], df["GAD7_Severity"], normalize="index")*100
gad_by_gender.plot(kind="bar", stacked=True, figsize=(7,5))
plt.title("GAD-7 Severity Distribution by Gender (%)")
plt.ylabel("Percentage")
plt.show()
```

```python
import networkx as nx

issue_cols = df.columns[6:38]
bin_df = df[issue_cols].fillna(0).astype(int)
co_mat = bin_df.T.dot(bin_df)

# Build graph
G = nx.Graph()
for i in co_mat.index:
    for j in co_mat.columns:
        if i != j and co_mat.loc[i,j] > 30:  # threshold to avoid clutter
            G.add_edge(i, j, weight=co_mat.loc[i,j])

plt.figure(figsize=(10,10))
pos = nx.spring_layout(G, k=0.5)
nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", edge_color="gray", font_size=8)
plt.title("Issue Co-occurrence Network")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

issue_cols = df.columns[6:38]
features = df[["PHQ-9 (Int)","GAD-7 (Int)"] + list(issue_cols)].fillna(0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=4, random_state=42)
df["Cluster"] = kmeans.fit_predict(X_scaled)

sns.scatterplot(data=df, x="PHQ-9 (Int)", y="GAD-7 (Int)", hue="Cluster", palette="tab10")
plt.title("Clusters of Students (Mental Health Profiles)")
plt.show()
```

```python
# Make sure Date column is datetime
df["Date (Int)"] = pd.to_datetime(df["Date (Int)"], errors="coerce")

# Select issue-related columns (adjust slice if needed)
issue_cols = df.columns[6:38]   # in your file, Family ... Health etc.
bin_df = df[issue_cols].fillna(0).astype(int)

# 1. Find top 5 issues overall
top5 = bin_df.sum().sort_values(ascending=False).head(5).index.tolist()

# 2. Aggregate by month
monthly = df.groupby(df["Date (Int)"].dt.to_period("M"))[top5].sum()

# Convert PeriodIndex back to timestamp for plotting
monthly.index = monthly.index.to_timestamp()

# 3. Plot
plt.figure(figsize=(10,6))
for col in top5:
    plt.plot(monthly.index, monthly[col], marker="o", label=col)

plt.title("Top 5 Issues Trend Over Time (Monthly)")
plt.xlabel("Month")
plt.ylabel("Number of Cases")
plt.legend()
plt.tight_layout()
plt.show()
```
